```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Benchmark of different implementation

```python
df1 = pd.read_csv('results.csv')
df1 = df1.sort_values(by='Matrix Size', ascending=True).round(3)
df1.head()
```

```python
df1.columns
```

```python
plt.plot(df1['Matrix Size'], df1['Threading'], linestyle='--', marker='o', label = 'Threading')
plt.plot(df1['Matrix Size'], df1['Multiprocessing'], marker= 'o', label= 'Multiprocessing')
plt.plot(df1['Matrix Size'], df1['Simple'], linestyle='-.', marker='o', label = 'Simple method')
plt.plot(df1['Matrix Size'], df1['C++ Code'], linestyle=':', marker='o', label = 'C++ method')
plt.xlabel('Matrix size')
plt.ylabel('Time')
plt.title('Comparaison of computation time by matrix size for different implementations')
plt.legend()
plt.show()
```
